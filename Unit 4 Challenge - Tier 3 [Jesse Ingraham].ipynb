**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
# function to calculate the average price ratio between two years
def create_price_ratio(df, b, y1, y2):
    """returns the ratio of the average price for y1 divided by the average price for y2"""
    
    avg_price_y1 = df.query('(London_Boroughs == @b) & (Year == @y1)')['Average_Price'].mean()
    avg_price_y2 = df.query('(London_Boroughs == @b) & (Year == @y2)')['Average_Price'].mean()
    
    return avg_price_y1 / avg_price_y2

# calculate price ratios for each London Borough between 2002 and 2022
price_ratios_dict = {}
for borough in properties.London_Boroughs.unique():
    price_ratios_dict[borough] = [create_price_ratio(properties, borough, 2002, 2022)]

# create dataframe containing price ratios
price_ratios_02_22 = pd.DataFrame(price_ratios_dict)
price_ratios_02_22 = price_ratios_02_22.transpose()
price_ratios_02_22 = price_ratios_02_22.rename(columns={0:'Price_Ratio_02_22'})
price_ratios_02_22 = price_ratios_02_22.sort_values('Price_Ratio_02_22')
price_ratios_02_22.reset_index()
```

```python
# Bonus: Plot each borough's average price as a function of time normalized by their earliest respective average prices

# ##########################
# ##   Import libraries   ##
# ##########################

# Let's import the pandas, numpy libraries as pd, and np respectively. 
import pandas as pd
import numpy as np

# Load the pyplot collection of functions from matplotlib, as plt 
from matplotlib import pyplot as plt

# #######################
# ##   Load the data   ##
# #######################

# London Average House Prices (lahp)
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"
lahp = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)

# ##########################
# ##   Wrangle the data   ##
# ##########################

# transpose the dataframe
lahp = lahp.transpose()

# reset the index
lahp = lahp.reset_index()

# assign the column headers
lahp.columns = lahp.iloc[0,:]

# remove the duplicate row
lahp = lahp.drop(0)

# rename bad column names
lahp = lahp.rename(columns={'Unnamed: 0':'London_Boroughs', pd.NaT:'ID'})

# remove the NaN rows
lahp = lahp.dropna(axis=0, how='any')

# remove the non-London Borough rows (using the reference: https://en.wikipedia.org/wiki/List_of_London_boroughs)
lahp = lahp.drop(lahp[lahp['London_Boroughs'].str.isupper()==True].index)
lahp = lahp.drop(lahp[lahp['London_Boroughs'].isin(['City of London', 'Inner London', 'Outer London', 'England'])].index)

# reset the index again
lahp = lahp.reset_index(drop=True)

# view new dataframe
lahp
```

```python
# melt dataframe to reorganize data
lahp = lahp.melt(id_vars = ['London_Boroughs', 'ID'], var_name = 'Date', value_name = 'Average_Price')

# sort by borough, date
lahp = lahp.sort_values(['London_Boroughs', 'Date'])

# reset the index again
lahp = lahp.reset_index(drop=True)

# add year column
lahp['Year'] = lahp['Date'].dt.year
```

```python
lahp.head()
```

```python
# for each borough, record the earliest price in a new series from a given year
u_boroughs = lahp.London_Boroughs.unique()
earliest_price_series = pd.Series(dtype='object')
norm_year = 2002
for borough in u_boroughs:
    earliest_date = lahp.query('(London_Boroughs == @borough) & (Year == @norm_year)')['Date'].min()
    earliest_price = lahp.query('(London_Boroughs == @borough) & (Date == @earliest_date)')['Average_Price']
    borough_rows = lahp.query('London_Boroughs == @borough').index
    series_to_add = pd.Series(np.repeat(earliest_price.to_numpy(), len(borough_rows)), index=borough_rows)
    earliest_price_series = pd.concat([earliest_price_series, series_to_add])
    
# add the series to the dataframe
lahp['Earliest_Price'] = earliest_price_series

# calculate the normalized average price per borough
lahp['Normalized_Average_Price'] = lahp['Average_Price'] / lahp['Earliest_Price']
```

```python
lahp.info()
```

```python
# #######################
# ##   Plot the data   ##
# #######################

# Pivot data for plotting
p_lahp = lahp.pivot(index='Date', columns='London_Boroughs', values='Normalized_Average_Price')

# Plot all data
ax = p_lahp.plot(grid=True)
ax.set_title("London Borough Home Price Growth (Normalized by 2002 Pricing)", color='black')
ax.set_ylabel("Normalized Average Price")
ax.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()

# Plot top 5 borough home price growths since January 2002
top_5_boroughs = p_lahp.tail(1).melt().sort_values('value').tail(5)['London_Boroughs']
top_5 = p_lahp[top_5_boroughs]
ax2 = top_5.plot(grid=True)
ax2.set_title("Top 5 London Borough Home Price Growth Since January 2002", color='black')
ax2.set_ylabel("Normalized Average Price")
ax2.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```
